```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
q1 = pd.read_csv("./query1.csv")
```

```python
q1.info
```

```python
q1
```

```python
u20 = q1[q1['year']==2020]["units_sold"].to_list()
u21 = q1[q1['year']==2021]["units_sold"].to_list()
```

```python
# 그림 사이즈, 바 굵기 조정
fig, ax = plt.subplots(figsize=(12,6))
bar_width = 0.4

# 판매 방식이 3개이므로 0, 1, 2 위치를 기준으로 삼음
index = np.arange(3)
method = q1.loc[0:2,"sales_method"]
t20=q1[q1['year']==2020]["total_sales"].to_list()
t21=q1[q1['year']==2021]["total_sales"].to_list()
# 각 연도별로 3개 샵의 bar를 순서대로 나타내는 과정, 각 그래프는 0.25의 간격을 두고 그려짐
b1 = plt.bar(index, t20, bar_width, alpha=0.4, color='blue', label='2020')

b2 = plt.bar(index + bar_width, t21, bar_width, alpha=0.4, color='red', label='2021')

# x축 위치를 정 가운데로 조정하고 x축의 텍스트를 method 정보와 매칭
plt.xticks(np.arange(bar_width/2, 3+bar_width/2), method)

# x축, y축 이름 및 범례 설정
plt.xlabel('method', size = 13)
plt.ylabel('total sales', size = 13)
plt.legend()
plt.show()
```

```python
print("연도별 매출")
for i in range(3):
    print(method[i],"비율:",f'{t21[i]/t20[i]:.0f}배')
```

```python
store_20 = q2.loc[(q2.sales_method=='In-store') & (q2.year==2020)]["price_per_unit"].to_list()
store_21 = q2.loc[(q2.sales_method=='In-store') & (q2.year==2021)]["price_per_unit"].to_list()
outlet_20 = q2.loc[(q2.sales_method=='Outlet') & (q2.year==2020)]["price_per_unit"].to_list()
outlet_21 = q2.loc[(q2.sales_method=='Outlet') & (q2.year==2021)]["price_per_unit"].to_list()
online_20 = q2.loc[(q2.sales_method=='Online') & (q2.year==2020)]["price_per_unit"].to_list()
online_21 = q2.loc[(q2.sales_method=='Online') & (q2.year==2021)]["price_per_unit"].to_list()
```

```python
plt.hist(store_20,alpha=0.4,color='blue',density=True,label="2020")
plt.hist(store_21,alpha=0.4,color='red',density=True,label="2021")
plt.legend()
plt.xlabel("Unit price",fontsize=15)
plt.ylabel("Relative frequency",fontsize=15)
plt.title("Distributuon of price in Store",fontsize=15)
```

```python
plt.hist(outlet_20,alpha=0.4,color='blue',density=True,label="2020")
plt.hist(outlet_21,alpha=0.4,color='red',density=True,label="2021")
plt.legend()
plt.xlabel("Unit price",fontsize=15)
plt.ylabel("Relative frequency",fontsize=15)
plt.title("Distributuon of price in Outlet",fontsize=15)
```

```python
plt.hist(online_20,alpha=0.4,color='blue',density=True,label="2020")
plt.hist(online_21,alpha=0.4,color='red',density=True,label="2021")
plt.legend()
plt.xlabel("Unit price",fontsize=15)
plt.ylabel("Relative frequency",fontsize=15)
plt.title("Distributuon of price in Online",fontsize=15)
```

```python
# 2020년 판매 방식에 따른 가격 히스토그램
plt.hist(store_20,alpha=0.4,color='r',density=True,label="Store")
plt.hist(outlet_20,alpha=0.4,color='g',density=True,label="Outlet")
plt.hist(online_20,alpha=0.4,color='b',density=True,label="Online")

plt.legend()
plt.xlabel("Unit price",fontsize=15)
plt.ylabel("Relative frequency",fontsize=15)
plt.title("Distributuon of price in 2020",fontsize=15)
```
